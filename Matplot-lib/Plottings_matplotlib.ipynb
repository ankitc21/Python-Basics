```python
import numpy as np
import matplotlib.pyplot as plt
x=np.array([1,6])
y=np.array([1,250])
plt.plot(x,y,'o')
plt.show()
```

```python
x=np.array([1,2,6,8])
y=np.array([3,8,1,10])
plt.plot(x,y)
plt.show()
```

```python
'''Default X-Points
If we do not specify the points on the x-axis, they will get the default values 0, 1, 2, 3 etc., depending on the length of the y-points.

So, if we take the same example as above, and leave out the x-points, the diagram will look like this:'''

y= np.array([3, 8, 1, 10, 5, 7])

plt.plot(y)
plt.show()
```

```python

y=np.array([3, 8, 1, 10])
plt.plot(y,marker='*')
plt.show()
```

```python

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, 'o:r')
plt.show()
```

```python

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, 'o:m')
plt.show()
```

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, 'o--k',ms=5)
plt.show()
```

```python
ypoints = np.array([3, 8, 1, 10])
plt.plot(ypoints,'o--r',mec='k',ms=10)
```

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints,'*:r', ms = 10, mfc = 'k')
plt.show()
```

```python
y=np.array([3, 8, 1, 10])
plt.plot(y, c='hotpink',linewidth = '20.5')
plt.show()
```

```python
x1 = np.array([0, 1, 2, 3])
y1 = np.array([3, 8, 1, 10])
x2 = np.array([0, 1, 2, 3])
y2 = np.array([6, 2, 7, 11])

plt.plot(x1, y1, x2, y2)
plt.show()
```
